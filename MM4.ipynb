## Resolução da questão 1-d):


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math

def a(n):
  return n**n/math.factorial(n)
ns = []
ans = []
n_min = int(input("Digite um valor para n_min: "))
n_max = int(input("Digite um valor para n_max: "))
for n in range(n_min, n_max+1):
    ns.append(n)
    ans.append(a(n))

#Tabela:
tabela = {'Valores de n' : ns, "Valores a(n)": ans}
tabela = pd.DataFrame(tabela)
display(tabela)

#Gráfico
sns.scatterplot(x=ns, y=ans,color='purple')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlabel('Valores de n')
plt.ylabel('Valores de a(n)')
plt.title('Gráfico de sequência númerica')
plt.show()
```

## Resolução da questão 2:
Questões a e b:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
n_min = int(input("Digite um valor para n_min: "))
n_max = int(input("Digite um valor para n_max: "))
Sns = []
values = []

for i in range(n_min, n_max+1):
    values.append(i)
    if i == n_min:  # Se for o primeiro:
        aux = i
        sn = 0
        while aux > 0:
            sn += 1 / aux
            aux = aux - 1
        Sns.append(sn)

    else:
        sn = Sns[-1] + 1/i
        Sns.append(sn)
#Tabela:
tabela = pd.DataFrame({"Valores de n" : values, "Valores de Sn":Sns})
display(tabela)
#Gráfico:
sns.scatterplot(x=values, y=Sns,color='purple')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlabel('Valores de n')
plt.ylabel('Valores de Sn')
plt.title('Gráfico de sequência Sn')
plt.show()
```

Questões c e d:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
k = 0.57721566
n_min = int(input("Digite um valor para n_min: "))
n_max = int(input("Digite um valor para n_max: "))
e = float(input("Digite um valor para ϵ: "))
ans = []
Sns = []
values = []
for i in range(n_min, n_max+1):
    values.append(i)
    if i == n_min:  # Se for o primeiro:
        aux = i
        sn = 0
        while aux > 0:
            sn += 1 / aux
            aux = aux - 1
        Sns.append(sn)

    else:
        sn = Sns[-1] + 1/i
        Sns.append(sn)
    ans.append(sn - math.log(i))
#Tabela:
tabela = pd.DataFrame({"Valores de n" : values, "Valores de Sn":Sns, "Valores de a(n)": ans})
display(tabela)
#Gráfico:
sns.scatterplot(x=values, y=ans,color='purple')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlabel('Valores de n')
plt.ylabel('Valores de a(n)')
plt.title('Gráfico com os valores de a(n)')
plt.axhline(y= k - e, color='red', linestyle='--', label='y = γ - ϵ')
plt.axhline(y= k , color='green', linestyle='--', label='y = γ')
plt.axhline(y= k + e, color='blue', linestyle='--', label='y = γ + ϵ')
plt.legend()
plt.show()
```
