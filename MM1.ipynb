Para isso eu calculei o limite da função escolhida resultando em 1, escolhi um ϵ arbitrário e calculei o menor número natural com essa tolerância N(ϵ) da seguinte forma -> Temos que o limite é 1, então: $$ | \frac{n}{n+1} - 1| \leq \epsilon$$
 Se o que tá dentro do módulo for positivo temos todos os valores possíves para n pois sempre n/n+1 - 1 < 0 e ϵ > 0. Dito isso, podemos calcular o N(ϵ) assim:$$ -\epsilon \leq \frac{n}{n+1} - 1  \Leftrightarrow \epsilon \geq  \frac{1}{n+1} \Leftrightarrow (n+1) \geq \frac{1}{\epsilon} \Leftrightarrow n \geq \frac{1}{\epsilon} - 1 $$
 Exemplo: substituindo ϵ por 0.01, temos que N(ϵ) = 9, ou seja tal aproximação é válido para todos os valores acima de 9, como observamos no seguinte cálculo:
$$n \geq \frac{1}{0.01} - 1 \Leftrightarrow n \geq 9$$
