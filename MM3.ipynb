No código abaixo o usuário escolhe os valores referentes as populações  das cidades A, B, C e o número de anos passados. Observa-se que sempre converge para A = 5000, B = 10000,  C = 20000.

```python
import pandas as pd

#Values é uma lista com os valores de A,B e C, escolhidos pelo usuário:
values = []
print(f"Indique valores para A,B e C, totalizando 35000\n")
values.append(float(input('A: ')))
values.append(float(input('B: ')))
values.append(float(input('C: ')))

def new_year(values):

  new_values = []
  new_values.append(0.9*values[0] + 0.05*values[2]) #An
  new_values.append(0.1*values[0]+ 0.8*values[1]) #Bn
  new_values.append(0.2*values[1] + 0.95*values[2]) #Cn

  return new_values

#O usuário escolhe quantos anos se passam:
years_passed = int(input("Quantos anos se passaram?"))

for i in range(1,years_passed+1):
  #Tabelas:
  new_values = new_year(values)
  print(f"Referente ao ano {i}° passado")
  tabela = {f'Valores ano {i-1}' : values, f"Valores ano {i}": new_values}
  tabela = pd.DataFrame(tabela)
  tabela.index = ['A', 'B', 'C']
  display(tabela)
  values = new_values
```
